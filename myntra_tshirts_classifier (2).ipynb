```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from keras.layers import BatchNormalization
import random
```

```python
image_directory='/content/drive/MyDrive/Myntra_Dataset/'
```

```python
df.index

```

```python
SIZE=224
X_dataset= []
r = list(range(100))
random.shuffle(r)
for i in tqdm(r):
  if i in df.index:
    img=image.load_img(image_directory+df['image_path'][i],target_size=(SIZE,SIZE,3))
    img=image.img_to_array(img)
    img=img/255
    X_dataset.append(img)
```

```python
X = np.array(X_dataset)
```

```python
Y=np.array(df.drop(["image_name","image_path"],axis=1))
```

```python
Y
```

```python
X_train, X_testval, Y_train, Y_testval = train_test_split(X, Y, random_state=20, test_size=0.3)
```

```python
X_test, X_val, Y_test, Y_val = train_test_split(X_testval, Y_testval, random_state=40, test_size=0.5)
```

```python
X_test.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from keras.optimizers import Adam
from keras.applications import EfficientNetB0
from keras import regularizers
SIZE = 224

class MyModel:
    def __init__(self, input_shape=(224, 224, 3), model_name='EfficientNetB0'):
        self.model_name = model_name
        self.base_model = None
        self.model = None

    def build_model(self, input_shape):
        # Load the EfficientNetB0 model
        self.base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape, pooling='avg')
        self.base_model.trainable=False
        # Add custom layers on top of the base model
        x = self.base_model.output
        x = BatchNormalization()(x)
        x = Dense(512, kernel_regularizer=regularizers.L2(0.01), activity_regularizer=regularizers.L1(0.005), bias_regularizer=regularizers.L1(0.005), activation='relu')(x)
        x = Dropout(0.3)(x)
        x = Dense(256, kernel_regularizer=regularizers.L2(0.01), activity_regularizer=regularizers.L1(0.005), bias_regularizer=regularizers.L1(0.005), activation='relu')(x)
        x = Dropout(0.4)(x)
        predictions = Dense(4, activation='sigmoid')(x)  # Adjust the number of units in the Dense layer to match your number of classes


        self.model = keras.Model(inputs=self.base_model.input, outputs=predictions)
        self.train_datagen = None
        self.val_datagen = None

    def compile_model(self, optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']):
        self.model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

    def create_datagen(self, X_train, X_val, batch_size=32):

        self.train_datagen = ImageDataGenerator(
            rescale=1./255,  # Normalize pixel values
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='nearest'
        )

        self.val_datagen = ImageDataGenerator(rescale=1./255)
        self.train_datagen.fit(X_train)
        self.val_datagen.fit(X_val)

    def fit_model(self, X_train, Y_train, X_val, Y_val, epochs=10):        #

        history = self.model.fit(self.train_datagen.flow(X_train, Y_train, batch_size=32),
                      epochs=epochs,
                      validation_data=self.val_datagen.flow(X_val, Y_val, batch_size=32),
                      validation_steps=None,verbose=1)
        return history

    def get_summary(self):
        return self.model.summary()

    def evaluate_model(self, X_test, Y_test):
        return self.model.evaluate(X_test, Y_test)
```

```python
model_instance = MyModel()
model_instance.build_model(input_shape=(224, 224, 3))
model_instance.compile_model()
model_instance.create_datagen(X_train, X_val)
```

```python
model_instance.get_summary()
```

```python
history = model_instance.fit_model(X_train, Y_train, X_val, Y_val,epochs=100)
```

```python
loss, accuracy = model_instance.evaluate_model(X_test, Y_test)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
